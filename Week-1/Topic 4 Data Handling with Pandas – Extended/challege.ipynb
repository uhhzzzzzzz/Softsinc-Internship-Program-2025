```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
print("First five records:")
display(titanic.head())
```

```python
print(titanic.isnull().sum())


titanic['age'] = titanic['age'].fillna(titanic['age'].median())

titanic = titanic.drop('deck', axis=1)

titanic = titanic.drop_duplicates()
print("\nCleaned data:", titanic.shape)
```

```python
survival_gender = titanic.groupby('sex')['survived'].mean()
survival_gender.plot(kind='bar', color=['pink', 'blue'])
plt.title('Survival rate by gender')
plt.ylabel('Survival rate')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(titanic['age'], bins=20, kde=True)
plt.title('Age distribution of passengers')
plt.xlabel('Age')
plt.show()
```

```python
sns.barplot(x='pclass', y='survived', data=titanic)
plt.title('Survival rate by travel status')
plt.xlabel('Class (1=First Class)')
plt.ylabel('Survival rate')
plt.show()
```

```python
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5)
plt.title('The relationship between age and rent')
plt.xlabel('Age')
plt.ylabel('Rent')
plt.show()
```

```python
sns.countplot(x='embarked', hue='survived', data=titanic)
plt.title('Boarding space and survival numbers')
plt.xlabel('Port')
plt.ylabel('Number')
plt.show()
```
